## Acceso a internet por velocidad y provincia

```python
AxsVel = pd.read_csv('/soyHENRY/PI03/Datasets/Internet_Accesos-por-velocidad.csv')
```

```python
AxsVel
```

```python
AxsVel.info()
```

```python
AxsVel['OTROS'].value_counts()
```

```python
# Rellenamos los NaN de la columna 'OTROS' con '0' (que es la mayoría de datos) para poder hacer uso de esas casillas.
AxsVel['OTROS'] =  AxsVel['OTROS'].fillna(0)
```

```python
AxsVel.info()
```

```python
#Se renombran las columnas ya que es un nombre muy largo y confuso (contiene caracteres especiales también).
AxsVel.rename(columns = {'HASTA 512 kbps':'AR1', '+ 512 Kbps - 1 Mbps':'AR2', '+ 1 Mbps - 6 Mbps':'AR3', '+ 6 Mbps - 10 Mbps':'AR4',
                         '+ 10 Mbps - 20 Mbps':'AR5', '+ 20 Mbps - 30 Mbps':'AR6', '+ 30 Mbps':'AR7'}, inplace = True)
```

```python
AxsVel
```

```python
AxsVel.to_csv('/soyHENRY/PI03/AxsVel.csv',index=False)
```
